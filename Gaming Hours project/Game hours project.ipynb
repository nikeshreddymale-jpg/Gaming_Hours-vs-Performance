Simple EDA Operations

```python
df.groupby("Game_Type")[["Academic_or_Work_Score", "Productivity_Level"]].mean()
```

Visualization

Daily Gaming Hours Distribution

```python
plt.figure()
plt.hist(df["Daily_Gaming_Hours"], bins=10)
plt.xlabel("Daily Gaming Hours")
plt.ylabel("Count")
plt.title("Distribution of Daily Gaming Hours")
plt.show()
```

Gaming Hours vs Academic Score

```python
plt.figure()
plt.scatter(df["Daily_Gaming_Hours"], df["Academic_or_Work_Score"])
plt.xlabel("Daily Gaming Hours")
plt.ylabel("Academic or Work Score")
plt.title("Gaming Hours vs Academic Performance")
plt.show()
```

Correlation Analysis

```python
# Correlation between numerical features
corr_matrix = df.select_dtypes(include="number").corr()
corr_matrix
```

Gaming Intensity Categorization

```python
# Create gaming intensity category
df["Gaming_Intensity"] = pd.cut(
    df["Daily_Gaming_Hours"],
    bins=[0, 2, 4, 6],
    labels=["Low", "Moderate", "High"]
)

df[["Daily_Gaming_Hours", "Gaming_Intensity"]].head()
```
